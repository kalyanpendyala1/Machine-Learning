# Dimensionality Reduction - 2

### PCA vs LDA

Both are used for Dimensionality Reduction.

PCA - Unsupervised Algorithm - Since it ignores the class labels. Its goal is to find direction of maximum variance and form PCs

LDA - Supervised algorithm - it Maximizes the separation between the classes.

PCA reduces dimensions by focusing on features with most variance, Whereas LDA is like PCA but it focuses on Maximizing separability among known categories.

PCA projects the data in direction of most variance, LDA projects data to a line that separates classes very well.

In PCA classes are mixed up, In LDA classes are well separated.

PCA is never used for classification but It can be used together with LDA to imporve the classification.

### LDA - Linear Discriminant Analysis

When reducing the 2D graph to 1D graph, if you ignore the y-axis and project all the points into x-axis, you are ignoring information provided by y-axis, similarly if you ignore x-axis and project it onto y-axis you are ignoring the information provided by the y-axis. Thats why it is bad idea.



LDA provides best solution, LDA uses information from both the axes to create a new axis, and it projects the data onto this new axis in a way to maximize the separation of two categories.



#### How does LDA creates new axis?

The new axis is created according to two criteria (Considered Simultaneously),

1. Maximize distance between means (d)

2. Minimize scatter within each category (s^2)



#### Why both distances and scatter are important?

If we only maximize the distance between the means without considering the scatter, the result is we get lot of overlap in the middle. This is not considered as the great separation.

However, if we optimize distance between means and the scatter, then we get nice separation.



#### What if we have more than two dimensions?

The process is the same, it creates an axis that maximizes the distance between the means while minimizing the scatter.

### What if we have more than two categories?

#### Step-by-step procedure for LDA

**Step1** - Calculate the distance between the mean of different classes. (between-class variance)

**Step2** - Calculate the distance between the mean and sample of every class   (witin-class variance).

**Step3** - Construct lower-dimensional space that maximizes Step1 (between-class variance) and minimizes Step2 (within-class variance).

### How to make predictions

LDA uses bayes' theorem for predictions, it estimates the probability that new instance belongs to every class. It assigns instance to a class that has most probability.

### Differences between SVM and LDA

In terms of classification, LDA finds the hyperplane that "best separates" **all** the data points while SVM looks for hyperplane that "best separates" **only** points between two classes.

LDA deals with means of two classes, while SVM mainly depends on the support vectors.

If the two classes are spherical then the LDA hyperplane is perpendicular to the line linking the two means of the points. LDA projection line is line perpendicular to separating hyperplane.




LDA assumes that data has some covariance and normally distributed.SVM has no such assumption.  Assumptions of SVM are margin should be as large as possible and support vectors are most useful points.

SVM is binary classifier. LDA is not

SVM chooses the classifier with the maximum margin, LDA finds mean of vectors of each class, the finds projection that maximizes separation of means.
