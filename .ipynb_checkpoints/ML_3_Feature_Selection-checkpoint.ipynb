## Feature Selection

### Hypothesis

Steps in Data Driven Decision Making

1. Formulate a Hypothesis

2. Find the right Test

3. Execute a test

4. Make a decision


An idea that can be tested.

*Apples in newyork are costly* It is an idea but it has no other thing to compare. It is not hypothesis.

*apples in newyork are costlyt 1.75 dollars* Accepted hypothesis.

*Mean salary of a Data Scientis is 113,000 dollars* Null hypothesis

anything other than null hypotheis is **Alternative Hypothesis**

h0 = salary = $ 113,000

h1 = salary != $ 113,000

h0 is commonly accepted fact, opposite is Alternate hypothesis. Researchers work to nullify or disapprove null hypothesis.

Null Hypothesis - Earth is flat.

Alternative Hypothesis - Earth is round.

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 
```

```python
data = pd.read_csv('mobile_price.csv')
X = data.iloc[:,0:20]
y = data.iloc[:,-1]

data.head()
```

```python
# apply SelectKBest class to extract top 10 best features

best_features = SelectKBest(score_func=chi2, k=10)
fit = best_features.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

dfscores
```

```python
# concat two dataframe for better visualization

featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Specs', 'Score']

featureScores.nlargest(10,'Score')
```

### Feature Importance

```python
from sklearn.tree import ExtraTreeClassifier

import matplotlib.pyplot as plt

model = ExtraTreeClassifier()
model.fit(X, y)

print(model.feature_importances_)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```
