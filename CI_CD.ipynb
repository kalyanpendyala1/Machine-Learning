## Ansible

It is open-source IT automation engine. It automates repetitive system administrative tasks. It uses YAML language in form of playbooks to describe automation jobs.

You can use Ansible to automate three types of tasks.

1. `Provisioning`: Set up the various servers you need in your infrastructure.

2. `Configuration Management`: Change the configuration of an application, OS, or device. Start or Stop services; Install or update applications; Implement security policy; perform wide variety of other configuration tasks.

3. `Application Deployment` : Automates the deployment of internally develope applications to your production systems.

#### Architecture



Ansible uses pull configuration. It works by connecting to your nodes and pushing out small programs called "Ansible Modules" to them. Ansible executes these modules over SSH and removes them when finished. 




Local Machine is where Ansible is installed and Nodes are systems to be configured. Nodes are connected to local machines through SSH clients. Module is collection of code files. These configurations files are called "Playbooks". 

Local Machine also has Inventory document that helps to group the nodes under labels. 

#### Playbook

playbooks are the instructions to configure the nodes. They are written in YAML, a language used to describe data. Playbook begins with three dashes (---), Playbook is list of plays. Host is the target for the play. Within each play we define which node we are targetting.

Each play is a list of tasks that we were looking to execute on nodes. Ex: Install apache task; Start Apache task; Install MySQL task etc.,

#### Inventory Management

It is a place where we maintain the structure of our network environment. An Inventory File classifies the nodes into groups.  When we want to execute our task we refer to the name of the groups and tasks will be executed on all the list of nodes of that group. 



#### Working of Ansible

Ansible is installed only on the local machine. This makes ansible agentless. It connects to differen nodes through SSH. The playbook and inventory are written at the local machine. 

Playbooks are sent to the nodes and executed on the remote hosts. This configures the nodes to their desired states. 

#### Ansible Tower

Ansible tower is a framework for Ansible. It provides GUI, Thus reducing the dependency on the command prompt window. Instead of typing long commands, tasks can now be performed in a single click.

# Docker

In Data Science we create models in the research environment but these codes may not work in the test environment due to difference in computer environment. So solution could be using Docker. 

Docker is very lightweight and provides functionality as virtual machines. But Docker containers are different from VMs, they occupy less space, Easy to scale and efficient.



#### What is Docker?

Docker is tool which is used to automate the deployment of applications in lightweight containers. So that application can work efficiently in different environments.

Container is a software package that consists of all dependencies required to run an application. Multiple containers run on same hardware and maintain isolated applications. 

Docker is used in situation where you have different software applications use different frameworks and you want to run all your applications with their software frameworks. Docker provides suitable environment for every application.



#### How does Docker work?

"Docker Engine" or Docker is base engine installed on your host machine to build and run containers using Docker components and services. It uses client-server architecture where client and server communicate using REST API. Docker client is service which runs a command, the command is translated using REST API and is sent to the docker Deamon (Server). Then Docker deamon checks the client request and interacts with Operating System in order to create or manage containers.



#### Components of Docker

4 components of Docker are,

1. Docker client and Server

2. Docker Images

3. Docker Containers

4. Docker Registry

**1.Docker Client and Server:**  Docker client is access from terminal and docker host runs the deamon and registry. A user can build docker images and run docker containers by passing commands from docker client to docker server.

**2.Docker Image:** Docker image is a template with instructions which is used for creating docker containers. A docker image is built using file called "Docker File". Docker image is stored in docker hub or in a repository so that other people can access it.

**3.Docker Container:** Docker container is a standalone executable software package which contains applications and their dependencies. Numerous docker containers run on same OS and share OS with other containers. Each container works in isolation. 


**4.Docker Registry:** Docker Registry is an OpenSource server side service used for hosting and distributing images. Docker has its own repository called Docker Hub. Here images can be stored in either public or private repositories. Public repositories can be used by anyone.

Push and Pull are commands used by users in order to interact with docker registry. In order to build a container, Pull command is used to get a docker image from the docker repository. A push command is used to store docker image in docker registry.



#### Docker Summary

1. Docker file creates a docker image using build command.

2. A docker image contains all the project code.

3. Using docker image any user can run the code in order to create docker containers.

4. Once a docker image is built, it is uploaded to docker registry or Docker Hub.

5. From the Docker Hub, users can get the Docker image and build new containers.
