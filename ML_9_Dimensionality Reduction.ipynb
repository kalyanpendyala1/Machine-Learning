### Why do we need to to Dimensionality Reduction

We perform dimensionality reduction because curse of dimensionality demands us to do.

Curse of dimensionality refers to all problems that arises when working with data in higher dimensions.

1. As the no. of features increases, we need to have proportional number of samples.

2. As the no. of features increases, model looks at complex patterns and overfits the data.

#### Advantages 

1. Less misleading data means, accuracy improves

2. Less computation and training time

3. Removes redundant features and noise

Dimensionality reduction removes inconsistencies, Redundant data and highly correlated features.

### Difference between feature selection and dimensionality reduction

They both aim at reducing the number of features in the original set of features. In feature selection we select features on the domain knowledge but what if we have around 300+ features?

Feature selection is simply selecting and excluding given features without changing them. But dimensionality reduction transforms features into a lower dimension.

The difference is that set of features made by feature selection must be a subset of the original set of features and set made by dimensionality reduction doesnt have to be subset, EX: PCA reduces dimensionality by making new synthetic features from covariances of the original ones, then discarding the less important ones.

Feature selection works on variance and dimensionality reduction works on Eigenvectors and Eigenvalues.

### Main approaches for Dimensionality Reduction

1. Projection

2. Manifold Learning

#### Projection



In most real-world problems, training instances are not spread uniformly across all dimensions. Many features are almost constant, while others are highly correlated.

As a result, all training instances actually lie within a much closer low dimensional "subspace" of high dimensional space.

If we project every training instance perpendicularly onto subspace we get 2D dataset. We reduced dimension from 3D to 2D. 

Projection is not always the best approach, if the dataset is int he shape of a swiss roll we cannot be able to project it. If we do so it will squash different layers of swiss-roll together.

#### Manifold Learning

https://towardsdatascience.com/dimensionality-reduction-pca-ica-and-manifold-learning-65393010253e

Swiss-roll is an example of manifold learning.

A 2D manifold is a 2D shape that can be bent and twisted in high-dimensional space. 

More generally, a d-dimensional manifold is a part of n-dimensional space (where d<n) that locally resembles a d-dimensional hyperplane. In case of swiss-roll d=2, n=3; it locally resembles a plane but it is rolled in 3rd Dimension.



Many dimensionality reduction algorithms work by modelling manifold on which training instances lie; This is called "Manifold Learning".

Manifold learning depends on manifold assumption or manifold hypothesis which holds that most real-world high-dimensional datasets lie close to much lower-dimensional manifold.

If you reduce dimensionality of the training set before training a model, it usually speedup the training, but it may not always lead to better or simple solution, all depends on the dataset.

### PCA (Principal Componenet Analysis)

First it identifies the hyperplane that lies closest to the data, and then it projects the data onto it, just like we did above.

It determines the directions along which the data has maximum variance. 

It returns 2-Dimensional planes if 3-Dimensional points are fed to PCA for evaluation, with thired dimension lying orthogonal to the plane. 

**Preserving the Variance**

Before you can project the training set into a lower-dimensional hyperplane, you first need to choose right hyperplane. *select the axis that preserves the maximum amount of variance*, as it will most likely to lose less information than the other projections.

our choice should *minimize the mean squared distance between the original dataset and its projection* onto that axis. 



a simple 2D dataset is represented on the left, along with three different axes(one dimensional hyperplanes). On the right is the result of the projection of the dataset onto each of these axes, as you can see the projection onto solid line preserves maximum variance.

Principal Component Analysis is Dimensionality Reduction technique that enables you to identify correlations and patterns in the data so that it can be transformed into a dataset of significantly lower dimensions without loosing much important information.

#### Why do we need to find correlation among different features?

Because highly correlated independent features usually cause an output which is really biased on those features. There might be some redundant features that do not account to output.

PCA finds highly correlated and redundant features and comes up with a new feature set where there is minimum correlation between features and maximum variation between features.

PCA doesnt take output variables into account.

#### PCA step by step

The below steps need to be followed to perform dimensionality reduction using PCA:

1. Standardization of the Data

2. Computing the covariance matrix

3. Calculating the eigenvectors and eigenvalues

4. Computing the principal components

5. Reducing the dimensions of the dataset.

#### STEP 1 - Standardation of the Data

Scaling your data such that all the values lie within a similar range.

Standardization is carried out by subtracting each value in the data from the mean and dividing it by the overall deviation in the dataset.



#### STEP 2 - Computing Co-variance matrix:

Covariance matrix represents the correlation between different variables in the dataset. It is essentially to identify heavily dependent variables because they contain biased and redundant information which reduces the overall performance of the model.




covariance matrix is PxP matrix where P represents no. of features. If you have 2 features you will get 2x2 matrix.

covariance denotes how these variables are dependent on each other. If covariance value is negative it represents both variables are inderectly proportional. If covariance value is positive then these both are directly proportional.

#### STEP 3 - Calculating eigenvectors and eigenvalues

Eigenvalues and eigenvectors are the mathematical constructs that must be computed fro the covariance matrix in order to determine principal components of the dataset.


Principal components are the new set of variables obtained from intial set  of variables. They compress and possess most of the useful information that was scattered among the initial variables.

Eigenvectors are those vectors when a linear transformation is performed on then then their direction doesnt change.

Eigenvalues are simply scalars of respective eigenvectors.

Principal components have to be highly significant when predicting the output and they also have to be very independent of each other.

All the data that is scattered around those 50 variables can be easily represented with these 5 variables which are known as principal components.

Initially you start by forming the same no. of principal components as your features in your dataset. The first principal component you formed should be in such a way that it stores the maximum possible information and similarly second principal component (PC2) should store remaining maximum information in your data and so on.. PC3 will be third most significant PC, PC4 will be 4th significant PC so on..

Then you arrange  PCs in descending orde, you know that first few PCs are most important and just take them and use them as new dataset.

#### What is the use of eigenvectors and eigenvalues in PCA?

The no. of features in your dataset tells you how many eigenvectors you need to create. If you have 2 features you will have two pairs of eigenvectors and eigenvalues.

The idea behind eigenvectors is to use covariance matrix to understand where in the data there is most amount of variance (Covariance matrix gives you overall variance among different variables)

Eigenvectors and eigenvalues are used to understand variance in your dataset. More variance in the dataset denotes more information about the data. Eigenvectors are used to identify where in your data you have maximum variance, in which direction, in which variable etc.,

The idea behind PCs, PCs are variables that store maximum information, maximum information is stored in variables where there is maximum variance.

#### STEP 4 - Computing Principal Components

Once we computed eigenvectors and eigenvalues, all we have to do is order them in descending order, where the eigenvector with highest eigenvalue is most significant and thus forms the first PC and so on.

#### STEP 5 - Reducing the dimensions of the dataset

rearrage dataset with final PCs

#### SVD to find Principal Components of a training set.

There is standard matrix factorization technique called "Singular Value Decomposition (SVD)", that can decompose the training matrix X into multiplication of matrix UΣV, Where v contains all the principal components we are looking for.

### Implementing PCA

https://sebastianraschka.com/Articles/2014_pca_step_by_step.html

```python
import numpy as np

np.random.seed(23)

mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])

# this method generates random samples
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20).T

# when the assertion fails it raises the error
assert class1_sample.shape == (3,20), "This matrix is not in the dimension of 3x20"
```

```python
mu_vec2 = np.array([0,0,0])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20).T
assert class2_sample.shape == (3,20), "This matrix is not in the dimension of 3x20"
```

using the code above we created two 3x20 datasets, one dataset for each class w1 and w2.

just to get rough idea how the samples of our two classes w1 and w2 are distributed, let us plot them in a 3d scatter plot.

```python
%pylab inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
plt.rcParams['legend.fontsize']=10
ax.plot(class1_sample[0,:], class1_sample[1,:], class1_sample[2,:], 'o', markersize=8, color='blue', alpha=0.5, label='class1')
ax.plot(class2_sample[0,:], class2_sample[1,:], class2_sample[2,:], '^', markersize=8, color='red', alpha=0.5, label='class2')


plt.title('Samples for class 1 and class 2')
ax.legend(loc='upper right')

plt.show()
```

#### 1. taking the whole dataset ignoring the class labels

because we dont need class labels for PCA analysis, let us merge the samples for our 2 classes into 3x40 dimensional array.

```python
all_samples = np.concatenate((class1_sample, class2_sample), axis=1)
assert all_samples.shape == (3,40), "The matrix has not the dimensions 3x40"
```

#### 2. Computing the d-dimensional mean vector

```python
mean_x = np.mean(all_samples[0,:])
mean_y = np.mean(all_samples[1,:])
mean_z = np.mean(all_samples[2,:])

mean_vector = np.array([[mean_x], [mean_y], [mean_z]])
mean_vector
```

#### 3.Computing the Covariance Matrix

```python
cov_mat = np.cov([all_samples[0,:],all_samples[1,:],all_samples[2,:]])
cov_mat
```

#### 4.computing eigenvectors and eigenvalues

```python
eig_val, eig_vec = np.linalg.eig(cov_mat)

for i in range(len(eig_val)):
    eigvec = eig_vec[:,i].reshape(1,3).T

print("Eigenvector {}: \n{}".format(i+1, eigvec[i]))
print("Eigenvalue {} from covariance matrix: {}".format(i+1, eig_val[i]))
```

#### 5.sorting eigenvectors by decreasing eigenvalues

```python
# make a list of (eigenvalue, eigenvector tuples)
eig_pairs = [(np.abs(eig_val[i]), eig_vec[:,i]) for i in range(len(eig_val))]

# sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# visually confirm that list is sorted correctly
for i in eig_pairs:
    print(i[0])
```

#### 5.2 Choosing k eigenvectors with the largest eigenvalues

```python
matrix_w = np.hstack((eig_pairs[0][1].reshape(3,1), eig_pairs[1][1].reshape(3,1)))
print('Matrix W:\n', matrix_w)
```

#### 6.Transforming the samples onto new subspace

```python
transformed = matrix_w.T.dot(all_samples)
assert transformed.shape == (2,40), "The matrix is not 2x40 dimensional"
```

```python
plt.plot(transformed[0,0:20], transformed[1,0:20], 'o', markersize=7, color='blue', alpha=0.5, label='class1')
plt.plot(transformed[0,20:40], transformed[1,20:40], '^', markersize=7, color='red', alpha=0.5, label='class2')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.title('Transformed samples with class labels')

plt.show()
```

### Best practices with scikit-learn PCA implementation

you can acess PCA from sklearn.decomposition class.

You can access Pricipal components using "components_" variable.

**How we can know how much variance lies in each principal component?**

'explained_variance_ratio' indicates the proportion of dataset variance that lies iin each of Principal Components.


**How to choose right number of PCs?**

We use 'n_components' parameter to specify no. of PCs we want.

Instead of choosing No. of dimensions to reduce down to, you can specify the 'n_components' to a float value between 0.0 to 1.0 indicating the ratio of variance you want to preserve. suppose you pass 0.95, it then computes
the minimum number of dimensions required to preserve 95% of the training set’s
variance.

#### Is it possible to decompress the reduced dataset?

Yes. We can do it by applying "Inverse Transformation" of the PCA projection. Ofcourse it wont give you exactly original data, since projection lost bit of information.

Mean Squared distance between reconstructed data and original data is called "Reconstruction Error".

In scikit-learn we use "inverse_transform()" method to do inverse transformation.

#### What is Randomized PCA?

By default svd_solver hyperparameter is set to auto. You can also set it to "randomized" to run Randomized PCA. When you set it, it uses stochastic algorithm called "Randomized PCA" that quickly finds an approximation of the first d principal components. IT is computationallly very fast. 

#### What is Incremental PCA?

Incremental PCA is implemented using "IncrementalPCA" class in scikit-learn.decomposition. You can split the large training set into mini-batches and feed IPCA one batch at a time.

It is useful for large training sets and also useful for implementing PCA in incremental learning.

Here we use partial_fit() method instead of fit(). fit() method is for full training set.

we use np.memap() method to split the training data.

#### What is Kernel PCA?

Kernel trick is mathematical trick that maps instances to higher dimensions to enable non-linear classification.

Linear decision boundary in higher-dimensional feature space corresponds to a complex non-linear decision boundary in original space.

Same trick can be applied to PCA to perform non-linear projections for dimensionality reduction. This is called Kernel PCA.

kernelPCA class can be imported from sklearn.decomposition.

####  How to select Kernel and Tuning Hyperparameters

You can simply use Grid search to select Kernel and hyperparameters that lead to best performance on the task. When you create a pipeline the first step reduces the dimensionality using Kernel PCA, then use SVR or Logistic Regression for classification.
